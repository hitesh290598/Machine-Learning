```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
bcancer = pd.read_csv('./Datasets/data.csv')
bcancer.head()
```

```python
bcancer = bcancer.drop(columns=['id', 'Unnamed: 32'])
```

```python
bcancer.head()
```

```python
le = preprocessing.LabelEncoder()
bcancer['diagnosis'] = le.fit_transform(bcancer['diagnosis'])
```

```python
X = bcancer.iloc[:,1:]
Y = bcancer.iloc[:,0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 0)
```

```python
print("X_train:");print(X_train)
print("\n\nX_test:");print(X_test)
print("\n\ny_train:");print(Y_train)
print("\n\ny_test:");print(Y_test)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=10, criterion = 'entropy', random_state=0)
rf_classifier.fit(X_train, Y_train)
```

```python
Y_pred = rf_classifier.predict(X_test)
Y_pred
```

```python
print('Classification Report:');print(metrics.classification_report(Y_test, Y_pred))
```

```python
print('Confusion Matrix');print(metrics.confusion_matrix(Y_test,Y_pred))
```

```python
print('Accuracy Score');print(metrics.accuracy_score(Y_test,Y_pred))
```

## Bagging

```python
dt_bg_clf = BaggingClassifier(DecisionTreeClassifier(random_state= 7), n_estimators=200, bootstrap=True, oob_score=True, random_state=42)
dt_bg_clf.fit(X_train, Y_train)
```

```python
y_pred = dt_bg_clf.predict(X_test)
y_pred
```

```python
print('Bagging Classifier Score:');print(metrics.accuracy_score(Y_test,y_pred))
```

```python
print('Confusion Matrix:');print(metrics.confusion_matrix(Y_test,y_pred))
```
